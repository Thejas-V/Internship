```python
# Importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('seaborn')
%matplotlib inline    
import warnings
warnings.filterwarnings("ignore")
import sklearn 
```

```python
# Finding dataset
df = pd.read_csv('customer_retention_dataset.csv')
df.head()
```

```python
df.tail()
```

```python
# Checking dataset
df.info()
```

```python
# finding null values
df.isna().sum().sum()
```

```python
sns.countplot(df['How old are you? '])
plt.show()
```

We see that the age group between 20-50 years use more of online services. This includes mostly working profesionals who are mostly busy and cannot visit the stores directly for shopping.

```python
from numpy import median
sns.barplot(x='How old are you? ', y='Since How Long You are Shopping Online ?',data=df, estimator=median)
plt.show()
```

We see that the age group between 20-50 years have used online services for more years than the people in other age group.

```python
sns.barplot(x='How many times you have made an online purchase in the past 1 year?', y='Since How Long You are Shopping Online ?',data=df, estimator=median)
plt.show()
```

The graph shows that the people who use online services more frequently have more chances of making a purchase. 

```python
sns.countplot(df['How much time do you explore the e- retail store before making a purchase decision?                                                                   '])
plt.show()
```

When people browse more then there is more tendency of purchasing a product. 

```python
sns.countplot(df['What is your preferred payment Option?					                                                                                                                               '])
plt.show()
```

The people who use credit/debit cards have the highest tendency to purchace a product. This happens due to easier access to money without much hassle transactions are done and the same can be told to people who use e-wallets and other modes of online payment. People who have money in hand i.e use cash on delivery are very less as they can easily go to stores and buy stuff instead of shopping online.

```python
sns.barplot(x='How frequently do you abandon (selecting an items and leaving without making payment) your shopping cart?							                                                   ', y='How much time do you explore the e- retail store before making a purchase decision?                                                                   ',data=df, estimator=median)
plt.show()
```

The people who take around 6-10 mins to make decision on what to purchase, dont abandon the cart much and proceed to buy the product. People who spend very less time or too much time to decide what to buy just add things to cart and cant decide whether to buy the product or not.

```python
sns.countplot(df['Why did you abandon the ?Bag?, ?Shopping Cart??					                                                                                                                                                       '])
plt.show()
```

The main reason for people to abandon the product they liked and added to cart, is they wait for better alternative or hope that better offers for the product pops up. 

```python
sns.barplot(x='Information on similar product to the one highlighted  is important for product comparison', y='How much time do you explore the e- retail store before making a purchase decision?                                                                   ',data=df, estimator=median)
plt.show()
```

The people who strongly agree that there need to be a similar product for comparison tend to spend more time before making a decision on buying a product. This leads to making them just keep searching for more better options than buy a product.

```python
fig,ax = plt.subplots(6,2, figsize=(15,40))
plt.suptitle("The distribution of key factors", fontsize=30)
sns.distplot(df['Ease of navigation in website'], ax = ax[0,0])
ax[0][0].set_title('Ease of navigation')
sns.distplot(df['Loading and processing speed'], ax = ax[0,1])
ax[0][1].set_title('Loading and processing speed')
sns.distplot(df['User friendly Interface of the website'], ax = ax[1,0])
ax[1][0].set_title('User friendly Interface')
sns.distplot(df['Convenient Payment methods'], ax = ax[1,1])
ax[1][1].set_title('Convenient Payment methods')
sns.distplot(df['Trust that the online retail store will fulfill its part of the transaction at the stipulated time'], ax = ax[2,0])
ax[2][0].set_title('Whether the time to deliver is same as given')
sns.distplot(df['Empathy (readiness to assist with queries) towards the customers'], ax = ax[2,1])
ax[2][1].set_title('Empathy towards the customers')
sns.distplot(df['Being able to guarantee the privacy of the customer'], ax = ax[3,0])
ax[3][0].set_title('Privacy of the customer surity')
sns.distplot(df['Responsiveness, availability of several communication channels (email, online rep, twitter, phone etc.)'], ax = ax[3,1])
ax[3][1].set_title('Responsiveness, availability of several communication channels')
sns.distplot(df['Online shopping gives monetary benefit and discounts'], ax = ax[4,0])
ax[4][0].set_title('Monetary benefit and discounts')
sns.distplot(df['Enjoyment is derived from shopping online'], ax = ax[4,1])
ax[4][1].set_title('Enjoyment from shopping online')
sns.distplot(df['Shopping online is convenient and flexible'], ax = ax[5,0])
ax[5][0].set_title('Shopping online is convenient and flexible')
sns.distplot(df['Return and replacement policy of the e-tailer is important for purchase decision'], ax = ax[5,1])
ax[5][1].set_title('Return and replacement policy')

plt.show()
```

The above plots are about the different factors that affect the repeated usage of a website by the customers. We can see that most of the customers strongly agree that what and how the website functions, the ease of access and use of website and its products is a major reason that makes it desirable for customers to shop online. Customers also agree that the website's interaction with them, how fast the returns can be got back and the benefits, discounts also plays a strong role for customer satisfaction.

```python
fig,ax = plt.subplots(2,2, figsize=(15,20))
sns.distplot(df['Shopping on the website gives you the sense of adventure'], ax = ax[0,0])
ax[0][0].set_title('Sense of adventure')
sns.distplot(df['Shopping on your preferred e-tailer enhances your social status'], ax = ax[0,1])
ax[0][1].set_title('Enhances your social status')
sns.distplot(df['Shopping on the website helps you fulfill certain roles'], ax = ax[1,0])
ax[1][0].set_title('Fulfill certain roles')
sns.distplot(df['Getting value for money spent'], ax = ax[1,1])
ax[1][1].set_title('Achieve value for money spent')

plt.show()
```

From the above plots we can see that the customers get an adventurous feeling which gives a sense of satisfation when they shop online. Most of the customers don't really think that shopping online increases their social status or helps them fulfill certain roles that they supposed to play, not do they reject the idea of it. But we can see that almost all the people who have purchased through online websites are satisfied about the fact that they get value of money they have spent.
