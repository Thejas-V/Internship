```python
#Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.

driver = webdriver.Chrome(r'C:\Users\DELL\Downloads\chromedriver_win32\Chromedriver.exe')
driver.get('https://www.naukri.com/')
designation=driver.find_element(By.CLASS_NAME,'suggestor-input')
designation.send_keys('Data Analyst')
location=driver.find_element(By.XPATH,"/html/body/div[1]/div[2]/div[3]/div/div/div[5]/div/div/div/input")
location.send_keys('Bangalore')
search=driver.find_element(By.CLASS_NAME,'qsbSubmit')
search.click()
```

```python
Job_title=[]
Job_location=[]
Company_name=[]
Exp_req=[]

Titles=driver.find_elements(By.XPATH,'//a[@class="title fw500 ellipsis"]')
for i in Titles[:10]:
    title=i.text
    Job_title.append(title)

Locations=driver.find_elements(By.XPATH,'//li[@class="fleft grey-text br2 placeHolderLi location"]')
for i in Locations[:10]:
    location=i.text
    Job_location.append(location)
    
Companies=driver.find_elements(By.XPATH,'//a[@class="subTitle ellipsis fleft"]')
for i in Companies[:10]:
    company=i.text
    Company_name.append(company)

Experience=driver.find_elements(By.XPATH,'//li[@class="fleft grey-text br2 placeHolderLi experience"]')
for i in Experience[:10]:
    exp=i.text
    Exp_req.append(exp)

df=pd.DataFrame({'Job_title':Job_title,'Job_location':Job_location,'Company_name':Company_name,'Experience_Req':Exp_req})
df
```

```python
# Q10: Write a python program to scrape the salary data for Data Scientist designation. You have to scrape Company name, Number of salaries, Average salary, Minsalary, Max Salary

driver = webdriver.Chrome(r'C:\Users\DELL\Downloads\chromedriver_win32\Chromedriver.exe')
driver.get('https://www.ambitionbox.com/')
time.sleep(3)
salaries=driver.find_element(By.XPATH,'/html/body/div[1]/nav[2]/div/ul/li[3]/a')
salaries.click()
time.sleep(3)
browse=driver.find_element(By.XPATH,'/html/body/div[1]/nav[2]/div/ul/li[3]/div/ul/li[1]/div')
browse.click()
time.sleep(3)
designation=driver.find_element(By.XPATH,'/html/body/div/div/div/main/section[1]/div[2]/div[1]/span/input')
designation.send_keys('Data Scientist')
time.sleep(3)
top=driver.find_element(By.XPATH,'/html/body/div/div/div/main/section[1]/div[2]/div[1]/span/div/div/div[1]/div/div')
top.click()
```

```python
Company_name=[]
Sal_Record=[]
Avg_Salary=[]
Max_Salary=[]
Min_Salary=[]
Exp_Req=[]

Companies=driver.find_elements(By.XPATH,'//div[@class="company-info"]')
for i in Companies[:10]:
    company=i.text.split("\n")[0]
    Company_name.append(company)

Record=driver.find_elements(By.XPATH,'//span[@class="datapoints"]')
for i in Record[:10]:
    rec=i.text
    Sal_Record.append(rec)
    
Average=driver.find_elements(By.XPATH,'//p[@class="averageCtc"]')
for i in Average[:10]:
    avg=i.text
    Avg_Salary.append(avg)

Minimum=driver.find_elements(By.XPATH,'//div[@class="salary-values"]/div[1]')
for i in Minimum[:10]:
    mn=i.text
    Min_Salary.append(mn)
    
Maximum=driver.find_elements(By.XPATH,'//div[@class="salary-values"]/div[2]')
for i in Maximum[:10]:
    mx=i.text
    Max_Salary.append(mx)
       
Experience=driver.find_elements(By.XPATH,'//div[@class="sbold-list-header"]')
for i in Experience[:10]:
    exp=i.text.split("(")[0]
    Exp_Req.append(exp)
    
df=pd.DataFrame({'Company_name':Company_name,'Sal_Record':Sal_Record,'Avg_Salary':Avg_Salary,'Min_Salary':Min_Salary,'Max_Salary':Max_Salary,"Exp_Req":Exp_Req})
df
```
