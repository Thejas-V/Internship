```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get("https://presidentofindia.nic.in/former-presidents.htm")
page
```

```python
content_of_page = BeautifulSoup(page.content)
content_of_page
```

```python
name = content_of_page.find("div",class_="presidentListing")
name.text
```

```python
presidents = []
for i in content_of_page.find_all("div",class_="presidentListing"):
    presidents.append(i.text)
    
presidents
```
