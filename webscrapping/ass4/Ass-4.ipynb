```python
# 2. Scrape the details team India‚Äôs international fixtures from bcci.tv

driver = webdriver.Chrome(r'C:\Users\DELL\Downloads\chromedriver_win32\Chromedriver.exe')
driver.get('https://www.bcci.tv/')
```

```python
International=driver.find_element(By.XPATH,"/html/body/nav/div[1]/div[2]/ul[1]/li[2]/a")
International.click()
```

```python
Series=[]
Titles=[]
Location=[]
Dates=[]
Time=[]

Ser=driver.find_elements(By.XPATH,'//span[@class="ng-binding"]')
for i in Ser:
    siri=i.text
    Series.append(siri)

Title=driver.find_elements(By.XPATH,'//span[@class="matchOrderText ng-binding ng-scope"]')
for i in Title:
    tit=i.text
    Titles.append(tit)
    
Locations=driver.find_elements(By.XPATH,'//div[@class="fix-place ng-binding ng-scope"]')
for i in Locations:
    loc=i.text.split("-")[1]
    Location.append(loc)
    
Date=driver.find_elements(By.XPATH,'//div[@class="match-card-left match-schedule"]')
for i in Date:
    D=i.text
    Dates.append(D)

times=driver.find_elements(By.XPATH,'//h5[@class="text-right ng-binding"]')
for i in times:
    T=i.text
    Time.append(T)
```

```python
df=pd.DataFrame({'Series':Series,'Titles':Titles,'Location':Location,'Dates':Dates,'Time':Time})
df
```

```python
# 5. Scrape the details of trending repositories on Github.com.

driver = webdriver.Chrome(r'C:\Users\DELL\Downloads\chromedriver_win32\Chromedriver.exe')
driver.get('https://github.com/')
```

```python
Open=driver.find_element(By.XPATH,'/html/body/div[1]/header/div/div[2]/div/nav/ul/li[3]/button')
Open.click()

Tr=driver.find_element(By.XPATH,'/html/body/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/ul[3]/li[3]/a')
Tr.click()
```

```python
Title=[]
Description=[]
Contributors=[]
Language=[]

Titles=driver.find_elements(By.XPATH,'//h1[@class="h3 lh-condensed"]')
for i in Titles[:10]:
    tit=i.text
    Title.append(tit)

Desc=driver.find_elements(By.XPATH,'//p[@class="col-9 color-fg-muted my-1 pr-4"]')
for i in Desc[:10]:
    D=i.text
    Description.append(D)
    
Contr=driver.find_elements(By.XPATH,'//a[@class="Link--muted d-inline-block mr-3"]')
for i in Contr[:10]:
    count=i.text
    Contributors.append(count)

Lang=driver.find_elements(By.XPATH,'//span[@class="d-inline-block ml-0 mr-3"]')
for i in Lang[:10]:
    lan=i.text
    Language.append(lan)
    
df=pd.DataFrame({'Title':Title,'Description':Description,'Contributors':Contributors,'Language':Language})
df
```
